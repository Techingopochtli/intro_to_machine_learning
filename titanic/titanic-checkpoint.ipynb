```python
#selection of columns with null values to determine how to deal with missing values
```

```python
cols_with_missing = [col for col in titanic_data.columns
                     if titanic_data[col].isnull().any()]
missing_val_count_by_column = (titanic_data.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
total_val_per_column = titanic_data["PassengerId"].count()
print(total_val_per_column)
```

```python
#Determine cardinality of categorical columns
```

```python
low_cardinality_cols = [cname for cname in titanic_data.columns if titanic_data[cname].nunique() < 10 and 
                        titanic_data[cname].dtype == "object"]

# Select numerical columns
numerical_cols = [cname for cname in titanic_data.columns if titanic_data[cname].dtype in ['int64', 'float64']]

print(low_cardinality_cols)
print(numerical_cols)
```
